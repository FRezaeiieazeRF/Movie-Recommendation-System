__importing librarys__

```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

__reading files(movies.csv and rating.csv) as data frames__

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
movies_df.head(10)
```

__removing the year part from the title and add a new column by using replace function__

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
#now the year column is with parentheses like (1995) 
#removing the parentheses from year column
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
#now the year column is fixed but the title column still has the year part in it.
#removing the year part from the title column by replaceing it with an empty string
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
#the title still has withspace at the ending of every row
#removing the whitespaces
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head(10)
```

# Content-Based
__geners column cant be used like this. we can solve this problem by turning the generes strings to a vector.<br> 
a vector of ones and zeros showing the movie has that gener or not.__

```python
#as genres column separats its generes by a | , we can call the split function on |
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_v2_df = movies_df.copy()
#by using a for(on the generes of the current row ) inside another for(on thw rows) we can make the vector we wanted
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        #wrting 1 at the column of gener we in the current row (to show the movie has this gener)
        movies_v2_df.at[index, genre] = 1
#filling the other cells(which means NAN cells) with 0 
movies_v2_df = movies_v2_df.fillna(0)
movies_v2_df.head()
```

```python
ratings_df.head()
```

```python
#as the time of when users rates does not matter. we drop the timestamp column
ratings_df = ratings_df.drop('timestamp', 1)
ratings_df.head()
```

__creating an input user to recommend movies__

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
#filtering the movies by their title
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
#merging it to get the movieId.
inputMovies = pd.merge(inputId, inputMovies)
#dropping useless information(year and geners) 
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)
#Final input dataframe of that person
inputMovies
```

```python
#filtering out the movies from the input
userMovies = movies_v2_df[movies_v2_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
#resetting the index to avoid future issues
userMovies = userMovies.reset_index(drop=True)
#dropping useless columns to creat gener table of the input user
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

```python
#dot input users rating with gener table(of the input user) to get weights
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
#the user gener interest
userProfile
```

```python
#getting the genres of every movie we have
genreTable = movies_v2_df.set_index(movies_v2_df['movieId'])
#droping the usless information
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable.head()
```

```python
#getting each movie score based in the input user
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
#sorting the recommendations in descending order
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head()
```

```python
#The final recommendation table
movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(10).keys())]
```

# Collaborative-Filtering

```python
#filtering out users that have watched the movies which the input has watched it too
userSubset = ratings_df[ratings_df['movieId'].isin(inputMovies['movieId'].tolist())]
userSubset.head(10)
```

```python
#group by the userid
userSubsetGroup = userSubset.groupby(['userId'])
```

```python
#sorting it so users with movie most in common with the input will have priority
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
```

```python
userSubsetGroup[0:3]
```

```python
#getting the first 50 users who has the most common with the input user
userSubsetGroup = userSubsetGroup[0:100]
```

```python
#a dictionary for storing Correlation
pearsonCorrelationDict = {}

#for the users whom we gathered to calculate the correlation
for usid, group in userSubsetGroup:
    #sorting the values (of the user) by movie id. (sorting is needed for calculating correlation)
    group = group.sort_values(by='movieId')
    #sorting the values (of the input user) by movie id. (sorting is needed for calculating correlation)
    inputMovies = inputMovies.sort_values(by='movieId')
    #number of the comnon movies which the user and input user has been rated is important to calculate correlation
    nRatings = len(group)
    #getting the rating of the user for the movies which both input user and the user have in common
    temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
    #list format is needed to facilitate future calculations
    tempRatingList = temp_df['rating'].tolist()
    tempGroupList = group['rating'].tolist()
    #calculating input user and the users correlation 
    Sxx = sum([i**2 for i in tempRatingList]) - (pow(sum(tempRatingList),2)/float(nRatings))
    Syy = sum([i**2 for i in tempGroupList]) - (pow(sum(tempGroupList),2)/float(nRatings))
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - (sum(tempRatingList)*sum(tempGroupList)/float(nRatings))
    #as if we cant divid by zero, an if is needed to check the values
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[usid] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[usid] = 0
```

```python
pearsonCorrelationDict.items()
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.columns = ['similarityIndex']
pearsonDF['userId'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head()
```

```python
#sorting the users by their simularity 
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]
topUsers.head()
```

```python
#getting all ratings which the selected users made 
topUsersRating=topUsers.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
topUsersRating.head()
```

```python
#we need to score the movies by the simularity of the seleceted users and the rate they have given to the movies
topUsersRating['weightedRating'] = topUsersRating['similarityIndex']*topUsersRating['rating']
topUsersRating.head()
```

```python
#we should calculate sum of the topUsers rating for all the movies. 
#calculating the sum of the simularity index of the users who rated, for each movie
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
tempTopUsersRating.head()
```

```python
#creating an empty dataframe for the recommendation
recommendation_df = pd.DataFrame()
#by dividing the sum_weightedRating by sum_similarityIndex we will get an score for each movie
recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
recommendation_df['movieId'] = tempTopUsersRating.index
recommendation_df.head()
```

```python
#sorting the top movies by their score
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_df.head(10)
```

```python
#the finall result of Collaborative Filterings recommendation
movies_df.loc[movies_df['movieId'].isin(recommendation_df.head(10)['movieId'].tolist())]
```
